```python
def call_whole_code():
    from tkinter import Tk, Frame, Label, Button, Entry, StringVar, IntVar, messagebox

    root = Tk()
    root.geometry("1350x750")
    root.title("PAN ATLANTIC CAFETERIA MENU")

    lbl_heading = Label(root, text="PAN ATLANTIC CAFETERIA MENU", font=('arial', 24, 'bold'))
    lbl_heading.pack(pady=20)

    BottomMainFrame = Frame(root, width=1350, height=650, bd=12, relief="raise")
    BottomMainFrame.pack(side="bottom")

    f1 = Frame(BottomMainFrame, width=450, height=650, bd=12, relief="raise")
    f1.pack(side="left")
    f2 = Frame(BottomMainFrame, width=450, height=650, bd=12, relief="raise")
    f2.pack(side="left")  
    f3 = Frame(BottomMainFrame, width=450, height=650, bd=12, relief="raise")
    f3.pack(side="left")

    menu_items = {
        "Jollof Rice": 350, "Coconut Fried Rice": 350, "Jollof Spaghetti": 350,
        "Sweet Chili Chicken": 1100, "Grilled Chicken Wings": 400, "Fried Beef": 400,
        "Fried Fish": 500, "Boiled Egg": 200, "Sauterated Sausages": 200,
        "Water": 200, "Glass Drink (35cl)": 150, "PET Drink (35cl)": 300, "PET Drink (50cl)": 350,
        "Glass/Canned Malt": 500, "Fresh Yo": 600, "Pineapple Juice": 350, "Mango Juice": 350,
        "Zobo Drink": 350, "Eba": 100, "Poundo Yam": 100, "Semo": 100,
        "Atama Soup": 450, "Egusi Soup": 450, "Savoury Beans": 350,
        "Roasted Sweet Potatoes": 300, "Fried Plantains": 150, "Mixed Vegetable Salad": 150,
        "Boiled Yam": 150
    }

    # Initialize variables to store item quantities
    item_quantities = {item: IntVar() for item in menu_items}

    def create_checkbox(frame, item, row, col):
        lbl_item = Label(frame, text=item, font=('arial', 12)) 
        lbl_item.grid(row=row, column=col, sticky="w")
        
        lbl_price = Label(frame, text=f"{menu_items[item]} Naira", font=('arial', 12))  
        lbl_price.grid(row=row, column=col+1, sticky="w")
        
        entry_quantity = Entry(frame, font=('arial', 12), width=4, justify='right', textvariable=item_quantities[item]) 
        entry_quantity.grid(row=row, column=col+2)

    rice_frame = Frame(f1, bd=12, relief="raise")
    rice_frame.pack(side="top", padx=10, pady=10)

    beverage_frame = Frame(f2, bd=12, relief="raise")  
    beverage_frame.pack(side="top", padx=10, pady=10) 

    side_dish_frame = Frame(f3, bd=12, relief="raise")  
    side_dish_frame.pack(side="top", padx=10, pady=10)  

    rice_items = ["Jollof Rice", "Coconut Fried Rice", "Jollof Spaghetti"]
    beverage_items = ["Water", "Glass Drink (35cl)", "PET Drink (35cl)", "PET Drink (50cl)", "Glass/Canned Malt", "Fresh Yo", "Pineapple Juice", "Mango Juice", "Zobo Drink"]
    side_dish_items = ["Savoury Beans", "Roasted Sweet Potatoes", "Fried Plantains", "Mixed Vegetable Salad", "Boiled Yam"]

    row = 0
    col = 0
    for item in rice_items:
        create_checkbox(rice_frame, item, row, col)
        row += 1

    row = 0
    col = 0
    for item in beverage_items:
        create_checkbox(beverage_frame, item, row, col)
        row += 1

    row = 0
    col = 0
    for item in side_dish_items:
        create_checkbox(side_dish_frame, item, row, col)
        row += 1

    varTotal = StringVar()  

    def food_price():
        total = 0
        for item, price in menu_items.items():
            quantity = item_quantities[item].get()
            total += quantity * price
        
        if total > 5000:
            total -= total * 0.25 
        elif 2500 <= total <= 5000:
            total -= total * 0.15 
        elif 1000 <= total < 2500:
            total -= total * 0.10  
        
        varTotal.set(total)

    btn_calculate_total = Button(f1, text="Calculate Total", command=food_price, font=('arial', 16, 'bold'))
    btn_calculate_total.pack(side="left", padx=10, pady=10)

    lbl_total = Label(f1, text="Total:", font=('arial', 16, 'bold'))
    lbl_total.pack(side="left", padx=5, pady=10)

    lbl_total_amount = Label(f1, textvariable=varTotal, font=('arial', 16, 'bold'))  # Use varTotal directly
    lbl_total_amount.pack(side="left", padx=5, pady=10)

    root.mainloop()

call_whole_code()
```

```python
import tkinter as tk
from tkinter import messagebox

root = tk.Tk()
root.title("PAU Cafeteria Menu")
root.geometry("800x800")

# Function to create labels for food items
def create_food_labels(frame, food_dict, x, y):
    for food, price in food_dict.items():
        food_label = tk.Label(frame, text=f"{food} \t\t\t{price}", font=("Helvetica", 8))
        food_label.place(x=x, y=y)
        y += 30

header_label = tk.Label(root, text="M E N U", font=("Helvetica", 32))
header_label.place(x=300, y=0)

# Group 1: RICE/PASTA
group1_label = tk.Label(root, text="RICE/PASTA", font=("Helvetica", 16))
group1_label.place(x=100, y=100)
rice_or_pasta = {"Jollof Rice": 350, "Coconut Fried Rice": 350, "Jollof Spaghetti": 350}
create_food_labels(root, rice_or_pasta, 20, 130)

# Group 2: SIDE DISHES
group2_label = tk.Label(root, text="SIDE DISHES", font=("Helvetica", 16))
group2_label.place(x=100, y=300)
side_dishes = {"Savoury Beans          ": 350, "Roasted Sweet Potatoes": 300, "Fried Plantains            ": 150, "Mixed Vegetable Salad": 150, "Boiled Yam                     ": 150}
create_food_labels(root, side_dishes, 20, 330)

# Group 3: SOUPS & SWALLOWS
group3_label = tk.Label(root, text="SOUPS & SWALLOWS", font=("Helvetica", 16))
group3_label.place(x=100, y=550)
soups_and_swallows = {"Eba          ": 100, "Poundo Yam": 100, "Semo             ": 100, "Atama Soup": 450, "Egusi Soup": 450}
create_food_labels(root, soups_and_swallows, 20, 580)

# Group 4: PROTEINS
group4_label = tk.Label(root, text="PROTEINS", font=("Helvetica", 16))
group4_label.place(x=500, y=100)
proteins = {"Sweet Chili Chicken": 1000, "Grilled Chicken Wings": 400, "Fried Beef                  ": 400, "Fried Fish                    ": 500, "Boiled Egg                 ": 200, "Sauteed Sausages    ": 200}
create_food_labels(root, proteins, 400, 130)

# Group 5: BEVERAGES
group5_label = tk.Label(root, text="BEVERAGES", font=("Helvetica", 16))
group5_label.place(x=500, y=400)
beverages = {"Water               ": 200, "Glass Drink (35cl)": 150, "PET Drink (35cl)": 300, "PET Drink (50cl)": 350, "Glass/Canned Malt": 500, "Fresh Yo": 600, "Pineapple Juice": 350, "Mango Juice": 350, "Zobo Drink": 350}
create_food_labels(root, beverages, 400, 430)

footer_label = tk.Label(root, text="ENJOY!", font=("Helvetica", 12))
footer_label.place(x=300, y=730)

button = tk.Button(root, text="Click here to place your order!", command=call_whole_code)

button.place(x=940, y=560)
button.config(fg="blue", bg="orange")

root.mainloop()
```
